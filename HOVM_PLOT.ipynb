```python
ds=xr.open_dataset('/home/eric/Desktop/Internshp/GROUP_2/DATA/cru_ts4.06.1901.2021.tmp.dat.nc.gz')
```

```python
dat_2_use=ds['tmp']
```

```python
NHA=dat_2_use.sel(lon=np.arange(-25,50,0.5), lat=np.arange(0,40,0.5), method ='nearest')
```

```python
NHA[0].plot()
```

```python
NHA.groupby('time.month').mean('time').mean(dim='lon').T.plot()
#I added T because I want the months to come to the x axis
```
