Subsetting 
=

```python
prep.where(prep['time.year']>2010,).dropna('time')# -999
```

```python
prep.where(prep['time.year']>2010,drop=True)# -999
```

```python
prep.where(prep['time.year']>2000,).where(prep['time.year']<2010,).dropna('time')# -999
#This gives the yearly values from 2001 to 2009 and dropped the others 
```

Grouping
=

```python
prep.groupby('time.month').mean('time').plot()# This grouped the data
#array into months, i.e from Jan. to Dec, calculated the mean and plotted.
```

```python
#To fine the monthly total
mon_total=prep.groupby('time.month').sum('time')
```

```python
#To calculate the yearly or annual total
Yr_total=prep.groupby('time.year').sum()
```

```python
#To calculate the daily total
D_total=prep.groupby('time.day').sum()
```

```python
ds_rh=xr.open_dataset('Relative_humidity_1981_2020.nc')
```

```python
ds_rh
```

```python
ds_rh.r.plot()
```

```python
ds_rh.r.mean('time').plot()# To plot the time mean of the data
```

```python
ds_rh.r.sel(longitude=0.5, latitude= 7.5, method='nearest').plot()
```

```python
ds_rh.r.mean('time').plot(cmap='RdBu')# To change the color, colors RdBu, 
# bwr, coolwarm,plasma, the _r is to reverse the color
```

```python
da_Ghana=ds_rh.r.sel(longitude=slice(-3.5,1.5), latitude=slice(11.5, 4))
```

```python
da_Ghana.mean('time').plot()# We founded the mean because it is a three dimensional data
```

```python
Ghana_mean=da_Ghana.mean('time')
```

```python
Ghana_mean.plot(cmap='twilight_shifted')
```

```python
# To find the yearly mean, convect the data to a yearly data

Year_rh=ds_rh.groupby('time.year').mean()
```

```python
Year_rh.r.plot()
```
