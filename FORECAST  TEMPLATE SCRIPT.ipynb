```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
import numpy as np
```

```python
file='W_C.xlsx'
```

```python
data = pd.read_excel(file)
```

# Export the data into a data frame

```python
df= pd.DataFrame(data)
```

```python
df
```

```python
df['HOURS']=pd.date_range(start='2022-10-28 12',end='2022-10-29 12',freq='3H')
df=df.set_index('HOURS')
```

```python
df
```

df

```python
g= df['TEMP']
F=df["RH"]
```

```python

fig, axes=plt.subplots(figsize=(8,5))


TIM=['12','1500','18','21','00','03','06','09','$12$']
# ax1=axes.twinx()
ax1=axes
ax1.plot(TIM,g,color='RED',lw=2, ls="-", marker="*")
ax1.set_xlabel('Time (UTC)',fontsize=15)
ax1.set_ylabel('Temperature($^oc$)',color='red', fontsize=15)

ax2=axes.twinx()
# ax2=axes
ax2.plot(TIM,F,color='BLUE',lw=2, ls="-", marker="o")
# ax2.set_xlabel('Hours (UTC)',fontsize=15)
ax2.set_ylabel('Relative Humidity (%)',color='blue',fontsize=15)
# ax2.set_title('Timeseries and bar chart',fontsize=15)
ax2.set_title('Temperature and Relative Humidity for Kumasi', fontsize=15)
fig.savefig("T_r.jpeg")

ax2.tick_params(axis='y', colors='blue')
ax1.tick_params(axis='y', colors='red')

```
