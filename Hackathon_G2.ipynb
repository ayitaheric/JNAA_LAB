```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
import pandas as pd
```

The Mean Temperature for the four stations
=

```python
Am=pd.read_csv('Accra_vectorcapacity_Tm.csv')
Km=pd.read_csv('Kumasi_vectorcapacity_Tm.csv')
Nm=pd.read_csv('Navrongo_vectorcapacity_Tm.csv')
Tm=pd.read_csv('Tamale_vectorcapacity_Tm.csv')
```

```python
AM=Am[["vc", "month"]].set_index("month")
KM=Km[["vc", "month"]].set_index("month")
NM=Nm[["vc", "month"]].set_index("month")
TM=Tm[["vc", "month"]].set_index("month")
```

```python
mon=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sept','Oct','Nov','Dec']
```

```python
plt.plot(mon,AM, label='Accra')
plt.plot(mon,KM, label='Kumasi')
plt.plot(mon,NM, label='Navrongo')
plt.plot(mon,TM, label='Tamale')
plt.legend(loc='lower left',bbox_to_anchor=(1,0.75))


plt.ylabel('Vectorial Capacity')
plt.xlabel('Month')
```

```python
fig,axes=plt.subplots(figsize=(18,14), ncols=2, nrows=2, linewidth=10)
plt.subplots_adjust(hspace = 0.4, wspace = 0.3)
x1 = axes[0,0]
x1.plot(mon,AN, label='Accra', marker='o')
x1.plot(mon,KN, label='Kumasi', marker='o')
x1.plot(mon,NN, label='Navrongo', marker='o')
x1.plot(mon,TN, label='Tamale', marker='o')
x1.set_title('Monthly Mean Minimun Temperature',fontsize=18)
x1.set_ylabel('Vectorial Capacity',fontsize=18)
x1.set_xlabel('Month',fontsize=18)
# x1.legend(loc='lower left',bbox_to_anchor=(1,0.70))

x2 = axes[0,1]
x2.plot(mon,AX,label='Accra', marker='o')
x2.plot(mon,KX, label='Kumasi', marker='o')
x2.plot(mon,NX,label='Navrongo', marker='o')
x2.plot(mon,TX, label='Tamale', marker='o')
# x2.legend(loc='lower left',bbox_to_anchor=(1,0.70))
x2.set_ylabel('Vectorial Capacity',fontsize=18)
x2.set_xlabel('Month',fontsize=18)
x2.set_title('Monthly Mean Maximum Temperature',fontsize=18)

x3 = axes[1,0]
x3.plot(mon,AM, label='Accra', marker='o')
x3.plot(mon,KM, label='Kumasi', marker='o')
x3.plot(mon,NM, label='Navrongo', marker='o')
x3.plot(mon,TM, label='Tamale', marker='o')
x3.legend(loc='lower left',bbox_to_anchor=(1,0.38),fontsize=30)#'xx-large'
x3.set_ylabel('Vectorial Capacity',fontsize=18)
x3.set_xlabel('Month',fontsize=18)
x3.set_title('Monthl Mean Temperature',fontsize=18)


x4=axes[1,1].set_visible(False)
plt.savefig('VC.tiff')
```

```python
Ap=An[["p", "tn"]].set_index("tn")
Kp=Kn[["p", "tn"]].set_index("tn")
Np=Nn[["p", "tn"]].set_index("tn")
Tp=Tn[["p", "tn"]].set_index("tn")
```

```python
plt.plot(Ap.index,Ap, label='Accra')
plt.plot(Kp.index,Kp, label='Kumasi')
plt.plot(Np.index,Np, label='Navrongo')
plt.plot(Tp.index,Tp, label='Tamale')
plt.legend(loc='lower left',bbox_to_anchor=(1,0.75))


plt.ylabel('Vectorial Survival')
plt.xlabel('Minimum Temperature')
```
